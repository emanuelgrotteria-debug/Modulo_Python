2. Si richiede di trovare per ogni continente:
a) il numero di casi fin
dall'inizio della pandemia
# b) la percentuale rispetto al totale del numero di casi


```python
# Prendiamo l'ultimo valore disponibile per ogni paese
last = cd.sort_values('date').groupby('location').tail(1)
#Sommiamo i casi per continente
continent_cases = last.groupby('continent')['total_cases'].sum()
#Calcoliamo il totale mondiale
world_total = continent_cases.sum()
#Percentuali per continente
continent_percentage = (continent_cases / world_total) * 100
#Mostriamo i risultati
print("Totale casi per continente:")
print(continent_cases)

print("\nPercentuale rispetto al totale mondiale:")
print(continent_percentage)

```

3. Selezionare i dati realtivi all'Italia nel 2022, poichè i dati vengono
registrati settimanalmente, filtrare via i giorni che non hanno misurazioni
quindi mostrare con dei grafici adeguati:
a) l'evoluzione dei casi totali dall'inizio alla fine dell'anno
b) il numero di nuovi casi rispetto alla data

```python
filtro_location = cd['location'] == 'Italy'
cd['date']= pd.to_datetime(cd['date'])
filtro_data =cd['date'].dt.year == 2022

filtro_nuovi_casi =cd['new_cases'] != 0.0
#tengo solo i giorni dove si registrano nuovi casi
filtro= filtro_location & filtro_data & filtro_nuovi_casi
Italia = cd[filtro]
Italia= Italia.loc[:,('iso_code',	'continent',	'location',	'date',
                      'total_cases', 'new_cases')]

Italia.plot(x = 'date', y= 'total_cases')
# visualizzazione andamento casi totali in Italia 2022


```

4. Riguardo le nazioni Italia e Germania

```python
filtro_data = cd['date'].between('2022-05-01', '2023-04-30')

filtro_nazioni = cd['location'].isin(['Italy', 'Germany', 'France'])

cd_filtrato = cd[filtro_data & filtro_nazioni]

cd_filtrato.boxplot(column='icu_patients', by='location') # boxplot con pandas
plt.title('Pazienti in terapia intensiva')
```

5. Riguardo le nazioni di Italia, Germania, Francia e Spagna in tutto il 2021:
a. mostrare, in maniera grafica oppure numerica, la somma dei pazienti ospitalizzati per ognuna (colonna hosp_patients)

b. se ci sono dati nulli, con un breve commento scrivere se può essere possibile gestirli tramite sostituzione o *meno*

```python
# Converte la colonna date in datetime
cd['date'] = pd.to_datetime(cd['date'])

# Filtra anno 2021
filtro_data1 = cd['date'].dt.year == 2021

# Paesi da confrontare
paesi = ['Italy', 'Germany', 'France', 'Spain']
cd_filtrato = cd['location'].isin(paesi) & filtro_data1

# Raggruppa e somma i pazienti ospedalizzati
cd_grouped = cd[cd_filtrato].groupby('location')['hosp_patients'].sum().round(0).astype(int)

cd_grouped
```

```python
#Rappresentazione grafica
cd_grouped.plot(kind= 'bar', x='location',y= 'hosp_patients' )
```
