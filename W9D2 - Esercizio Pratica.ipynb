```python
#Esercizio 11
guadagni = [100, 90, 70, 40, 50, 80, 90, 120, 80, 20, 50, 50]

somma = 0
for valore in guadagni:
    somma += valore

media = somma / len(guadagni)

print("La media dei guadagni è:", media)
```

```python
#Esercizio 12
parole = ["Albergo", "Sedia", "Borgo", "Petalo", "Eremo", "Belvedere",
          "Semestre", "Esteta", "Sosta", "Orpello", "Abete", 
          "Orologio", "Cesta", "Ermellino"]
for parola in parole:
    #trasformo la parola in minuscolo
    parola_minuscola = parola.lower()
     #conto le "e"
    conteggio = parola_minuscola.count("e")
    
    print(parola, "→ contiene", conteggio, "e")
```
